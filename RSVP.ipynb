```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("RSVP.csv")
df.head()
```

```python
df = df.drop("Timestamp",axis = 1)
df.head()
```

```python
df.isna().sum()
```

```python
# Mean
# average daily sleep
mean_sleep = np.mean(df["On average, how many hours of sleep do you typically get on weekdays?"])
print(f"On average EUI students sleeps for {int(mean_sleep)+1} hours on weekdays.")
# average caffeine drink consuption
mean_caffeine = np.mean(df["How often do you typically have any caffeine drink in a day(tea,coffee,etc)?"])
print(f"On average EUI students consume around {int(mean_caffeine)} drinks per day.")
# Median
median_sleep = np.median(df["On average, how many hours of sleep do you typically get on weekdays?"])
print(f"EUI students have a sleep median of {median_sleep} hours on weekdays.")
# average caffeine drink consuption
median_caffeine = np.median(df["How often do you typically have any caffeine drink in a day(tea,coffee,etc)?"])
print(f"EUI students median consumption of caffeine drinks is around {int(median_caffeine)} drink per day.")
# Mode
mode_sleep = stats.mode(df["On average, how many hours of sleep do you typically get on weekdays?"])
print(f"{mode_sleep[1]} EUI students sleeps for {mode_sleep[0]} hours per day.")
mode_caffeine = stats.mode(df["How often do you typically have any caffeine drink in a day(tea,coffee,etc)?"])
print(f"{mode_caffeine[1]} EUI students drinks {mode_caffeine[0]} cups per day.")
```

```python
academic_rating_as_int = []
for index in range(len(df['How much would you rate your academic performance according to your grades '])):
    if df['How much would you rate your academic performance according to your grades '][index] == 'Bad':
        academic_rating_as_int.append(1)
    elif df['How much would you rate your academic performance according to your grades '][index] == 'Neutral':
        academic_rating_as_int.append(2)
    elif df['How much would you rate your academic performance according to your grades '][index] == 'Good':
        academic_rating_as_int.append(2)
    elif df['How much would you rate your academic performance according to your grades '][index] == 'Very Good':
        academic_rating_as_int.append(2)
    elif df['How much would you rate your academic performance according to your grades '][index] == "Excellent":
        academic_rating_as_int.append(2)
academic_rating_as_int = np.array(academic_rating_as_int)
```

```python
academic_rating_as_int
```

```python
np.corrcoef(academic_rating_as_int,df['On average, how many hours of sleep do you typically get on weekdays?'])[1][0]
```

```python
# Scatter Plot between C-GPA and Preparation
plt.scatter(df['On average, how many hours of sleep do you typically get on weekdays?'],academic_rating_as_int,color='coral')
plt.xlabel("Hours Of Sleep")
plt.ylabel("Academic Rating")
plt.title("Scatter Plot of Sleep and Academic Level")
plt.grid(True, linestyle='-', alpha=0.5)
plt.show()
```
